```python
import matplotlib.pyplot as plt

f = open('대기오염_20241201223221.csv', encoding='utf-8') 
data = csv.reader(f)
header = next(data) 

month = []
result = []

for row in data:
    try:
        if row[0].strip() == "" or row[1].strip() == "" or "미세먼지" in row[1]:
            continue
        month.append(row[0])  
        result.append(float(row[1]))  
    except ValueError as e:
        print(f"데이터 수정 중 오류 발생: {e}, 데이터: {row}")
        continue

f.close()

plt.close()
plt.title('서울시 월별 미세먼지 농도 변화')
plt.scatter(month, result, color='r')
plt.xlabel('월별')
plt.ylabel('미세먼지 농도 변화')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('대기오염_20241201223221.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data) 

years = [str(year) for year in range(2012, 2023)] 
monthly_data = {year: [] for year in years}  

for row in data:
    month = row[0]
    for i, year in enumerate(years):
        try:
            monthly_data[year].append(float(row[i + 1])) 
        except ValueError:
            continue

f.close()

avg_pm_values = [sum(monthly_data[year]) / len(monthly_data[year]) for year in years]

plt.figure(figsize=(10, 6))
plt.bar(years, avg_pm_values, color='skyblue')

plt.title('서울시 년도별 미세먼지 농도 (평균)')
plt.xlabel('년도')
plt.ylabel('평균 미세먼지 농도 (㎍/㎥)')

plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('대기오염(구별)_20241201222607.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data) 

districts = []
avg_pm10_values = []

for row in data:
    try:
        district = row[0]
        
        pm10_values = [float(value) for value in row[1:]]  

        avg_pm10 = sum(pm10_values) / len(pm10_values)

        districts.append(district)
        avg_pm10_values.append(avg_pm10)

    except ValueError as e:
        print(f"데이터 수정 중 오류 발생: {e}, 데이터: {row}")
        continue

f.close()

plt.figure(figsize=(10, 6))
plt.bar(districts, avg_pm10_values, color='skyblue')

plt.title('서울시 구별 미세먼지 농도 (2015-2022년 평균)')
plt.xlabel('구별')
plt.ylabel('평균 미세먼지 농도 (㎍/㎥)')

plt.xticks(rotation=90)

plt.tight_layout() 
plt.show()
```
