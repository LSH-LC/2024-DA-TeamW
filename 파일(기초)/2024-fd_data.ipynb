- 그래프 그려보기

```python
import csv
import datetime
import matplotlib.pyplot as plt

f = open('기온 2019~2024.csv', encoding = 'EUC-KR')
data = csv.reader(f)

header = next(data)

x_result = []
y_result = []
for row in data:
    y_result.append(float(row[2]))
    x_result.append(datetime.datetime.strptime(row[0], '%b-%y'))
    
f.close()

plt.plot(x_result, y_result)
plt.xlabel('year-month')
plt.ylabel('temperature')
plt.grid(True)
plt.show()
```

## 그래프 합치기

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 첫 번째 데이터: 미세먼지 (PM)
f1 = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data1 = csv.reader(f1)

header1 = next(data1)

x_result = []
y_result_pm = []  # 미세먼지 데이터 저장

for row in data1:
    y_result_pm.append(float(row[1]))
    x_result.append(datetime.datetime.strptime(row[0], '%b-%y'))

f1.close()

# 두 번째 데이터: 강수량 (Rainfall)
f2 = open('강수량 2019~2024.csv', encoding='EUC-KR')
data2 = csv.reader(f2)

header2 = next(data2)

y_result_rainfall = []  # 강수량 데이터 저장

for row in data2:
    y_result_rainfall.append(float(row[2]))

f2.close()

# 그래프 합치기
plt.plot(x_result, y_result_pm, label='PM (Fine Dust)', color='blue')  # 미세먼지 그래프
plt.plot(x_result, y_result_rainfall, label='Rainfall', color='green')  # 강수량 그래프

# 레이블 추가
plt.xlabel('Year-Month')
plt.ylabel('Value')
plt.title('PM and Rainfall over Time')
plt.grid(True)

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```

- 그래프 크기를 같게 하기

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 첫 번째 데이터: 미세먼지 (PM)
f1 = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data1 = csv.reader(f1)

header1 = next(data1)

x_result = []
y_result_pm = []  # 미세먼지 데이터 저장

for row in data1:
    y_result_pm.append(float(row[1]))
    x_result.append(datetime.datetime.strptime(row[0], '%b-%y'))

f1.close()

# 두 번째 데이터: 강수량 (Rainfall)
f2 = open('강수량 2019~2024.csv', encoding='EUC-KR')
data2 = csv.reader(f2)

header2 = next(data2)

y_result_rainfall = []  # 강수량 데이터 저장

for row in data2:
    y_result_rainfall.append(float(row[2]))

f2.close()

# 그래프 설정
fig, ax1 = plt.subplots()

# 첫 번째 y축 (미세먼지 - 갈색 선)
ax1.plot(x_result, y_result_pm, label='PM (Fine Dust)', color='brown')
ax1.set_xlabel('Year-Month')
ax1.set_ylabel('PM (Fine Dust)', color='brown')
ax1.tick_params(axis='y', labelcolor='brown')

# 두 번째 y축 (강수량 - 파란색 선)
ax2 = ax1.twinx()  # 두 번째 y축 추가
ax2.plot(x_result, y_result_rainfall, label='Rainfall', color='blue')
ax2.set_ylabel('Rainfall (mm)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# 제목 추가
plt.title('PM and Rainfall over Time')

# 그래프에 범례 추가
fig.tight_layout()  # 레이아웃 조정 (겹침 방지)

# 그래프 표시
plt.show()
```

- 비교하기 적당하게 크기조절 한 그래프

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 첫 번째 데이터: 미세먼지 (PM)
f1 = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data1 = csv.reader(f1)

header1 = next(data1)

x_result = []
y_result_pm = []  # 미세먼지 데이터 저장

for row in data1:
    y_result_pm.append(float(row[1]))
    x_result.append(datetime.datetime.strptime(row[0], '%b-%y'))

f1.close()

# 두 번째 데이터: 강수량 (Rainfall)
f2 = open('강수량 2019~2024.csv', encoding='EUC-KR')
data2 = csv.reader(f2)

header2 = next(data2)

y_result_rainfall = []  # 강수량 데이터 저장

for row in data2:
    y_result_rainfall.append(float(row[2]))

f2.close()

# 그래프 설정
fig, ax1 = plt.subplots()

# 첫 번째 y축 (미세먼지 - 갈색 선)
ax1.plot(x_result, y_result_pm, label='PM (Fine Dust)', color='brown')
ax1.set_xlabel('Year-Month')
ax1.set_ylabel('PM (Fine Dust)', color='brown')
ax1.tick_params(axis='y', labelcolor='brown')

# ax1의 y축 범위 축소
# 예를 들어, 범위를 y_result_pm의 최소값 90%와 최대값 110%로 설정하여 축소
ax1.set_ylim(min(y_result_pm) * 0.25, max(y_result_pm) * 1.75)

# 두 번째 y축 (강수량 - 파란색 선)
ax2 = ax1.twinx()  # 두 번째 y축 추가
ax2.plot(x_result, y_result_rainfall, label='Rainfall', color='blue')
ax2.set_ylabel('Rainfall (mm)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# 제목 추가
plt.title('PM and Rainfall over Time')

# 그래프에 범례 추가
fig.tight_layout()  # 레이아웃 조정 (겹침 방지)

# 그래프 표시
plt.show()
```
